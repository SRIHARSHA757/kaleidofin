Q3)Write an R/Python script which calculates the annualized return of chit fund participant ?- Show the Return % for each month's bid winner.

```python
def cal_anualized_return_of_participants(df):
    
    participant_anualized_return=[]
    return_percent_of_each_month_bid_winner=[]
    person_spent_for_every_month=2000
    
    for i in range(len(df)):
        # i th month bid winner net return on i th month
        ith_month_return=df.iloc[i,4]
        #total amount recieved by the bid winner by amount returned to everyone in the group
        total_return_by_amount_returned_to_everyone=df.iloc[:25,5].sum()
        # total amount get by the i th person after 25 months
        total_return=ith_month_return+total_return_by_amount_returned_to_everyone
        #total amount spent by the every person in the span of 25 months
        total_spent_amount=25*2000
        #anualized_return of the person i th person
        anualized_return=total_return-total_spent_amount
        participant_anualized_return.append(anualized_return)
        #Return % for ith month's bid winner
        return_percent=(anualized_return/person_spent_for_every_month)*100
        return_percent_of_each_month_bid_winner.append(return_percent)
        
    for j in range(len(return_percent_of_each_month_bid_winner)):
        return_percent_of_each_month_bid_winner[j]=str(return_percent_of_each_month_bid_winner[j])+'%'
        
    a=np.array(participant_anualized_return)
    a=np.expand_dims(a,axis=1)
    b=np.array(return_percent_of_each_month_bid_winner)
    b=np.expand_dims(b,axis=1)

    c=np.concatenate((a,b),axis=1)
    column=["anualized_return_of_person(bid_winner_of_that_month)","percentage_return_of_each_month_bid_winner"]
    data_frame=pd.DataFrame(c,columns=column)

    return data_frame
```

```python
cal_anualized_return_of_participants(df)
```
